# Crime Analysis Project

Importing libraries

```python
import pandas as pd
import datetime
```

Loading data

```python
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv')
pd.set_option('display.max_columns',30)
```

```python
df.head()
```

```python
# infomation of data

df.info()
```

```python
# Removing unnessary rows(which is not usefull for analysis)

df.drop(columns=['AREA','Rpt Dist No','Mocodes','Premis Cd','Weapon Used Cd','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4','Cross Street','LAT','LON','Crm Cd'],inplace=True)
```

```python
df.sample(5)
```

```python
df.info()
```

```python
# Converting Timestamp to data only(means removing time from time stamp)

df['DATE OCC'] = df['DATE OCC'].str.split(' ').str.get(0)
df['Date Rptd'] = df['Date Rptd'].str.split(' ').str.get(0)
```

```python
# converting date from object to datetime

df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
```

```python
# mapping part-1 and part-2 with Major and Minor Crime respectively

df['Part 1-2'] = df['Part 1-2'].astype('str')
df['Part 1-2'] = df['Part 1-2'].map({'1':'Major Crime','2':'Minor Crime'})
```

Filling missing values

```python
df['Vict Sex'] = df['Vict Sex'].fillna('Unknown')
```

```python
df['Vict Descent'] = df['Vict Descent'].fillna('Other')
```

```python
df['Premis Desc'] = df['Premis Desc'].fillna('STREET')
```

```python
df['Weapon Desc'] = df['Weapon Desc'].fillna('No Weapon')
```

```python
df['TIME OCC'] = df['TIME OCC'].astype('str').str.zfill(4)
```

```python
# extracting hour and minute from date column

df['Hour'] = df['TIME OCC'].str[0:2].astype(int)
df['Minute'] = df['TIME OCC'].str[2:].astype(int)
```

```python
df.head()
```

```python
# Here also we do mapping

df['Vict Sex'] = df['Vict Sex'].replace({'X':'Unknown','H':'Unknown','-':'Unknown','M':'Male','F':'Female'})
```

```python
descent_map = {
    'A': 'Asian',
    'B': 'Black',
    'C': 'Chinese',
    'D': 'Cambodian',
    'F': 'Filipino',
    'G': 'Guamanian',
    'H': 'Hispanic/Latino',
    'I': 'American Indian/Alaskan Native',
    'J': 'Japanese',
    'K': 'Korean',
    'L': 'Laotian',
    'O': 'Other',
    'P': 'Pacific Islander',
    'S': 'Samoan',
    'U': 'Hawaiian',
    'V': 'Vietnamese',
    'W': 'White',
    'X': 'Unknown',
    'Z': 'Asian Indian'
}

df['Vict Descent'] = df['Vict Descent'].map(descent_map)
```

```python
# removing invalid dates

df = df[(df['Vict Age'] >= 1) & (df['Vict Age'] <= 100)]
```

```python
# making age group from age column 

def get_age_group(age):
    if age <= 10:
        return 'Child'
    elif age <= 17:
        return 'Teen'
    elif age <= 25:
        return 'Youth'
    elif age <= 40:
        return 'Adult'
    elif age <= 60:
        return 'Middle Age'
    elif age <= 80:
        return 'Senior'
    else:
        return 'Elderly'

df['age_group'] = df['Vict Age'].apply(get_age_group)
```

```python
# saving cleaned data

df.to_csv('cleaned_crime_data.csv',index=False)
```
